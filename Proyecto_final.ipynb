```python
# URL de la página 
url = "https://fbref.com/en/comps/Big5/Big-5-European-Leagues-Stats"
# Realizar una solicitud GET a la página
response = requests.get(url)
# Comprobar si la solicitud fue exitosa
if response.status_code == 200:
    # Parsear el contenido HTML de la página con Beautiful Soup
    soup = BeautifulSoup(response.text, 'html.parser')
    # Encontrar todos los elementos "tbody" en la página
    tbody_elements = soup.find_all('tbody')
    # Iterar a través de los elementos "tbody"
   
    for tbody in tbody_elements:
        # Encontrar todos los elementos "td" con el atributo "data-stat" igual a "team" dentro del "tbody"
        team_td_elements = tbody.find_all('td', attrs={"data-stat": ["team"]})
        
        # Iterar a través de los elementos "td" y obtener el texto de cada uno
        for td_element in team_td_elements:
            team_list = td_element.get_text(strip=True) # Obtener el texto del elemento sin espacios en blanco innecesarios
            print(team_list)   
else:
    print("La solicitud no fue exitosa. Código de estado:", response.status_code)
```

```python
#Primero comprobaremos que el codigo funciona y podemos extraer los links de cada equipo en una lista
url = "https://fbref.com/en/comps/Big5/Big-5-European-Leagues-Stats"

# Realizar la solicitud GET a la página web
response = requests.get(url)

# Verificar si la solicitud fue exitosa (código de estado 200)
if response.status_code == 200:
    # Parsear el contenido HTML con BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Encontrar todos los enlaces dentro del tag 'tbody' que están dentro de un td con data-stat="team"
    tbody_links = soup.select('tbody td[data-stat="team"] a[href]')

    # Extraer los enlaces y mostrarlos
    for link in tbody_links:
        href = link['href']
        print(href)
else:
    print(f"Error al realizar la solicitud. Código de estado: {response.status_code}")
```

```python
#Una vez comprobado que el codigo si funciona, procedemos a guardar los links dentro de una lista "links_equipos"
soup = BeautifulSoup(response.text, 'html.parser')
# Encontrar todos los enlaces dentro del tag 'tbody' que están dentro de un td con data-stat="team"
tbody_links = soup.select('tbody td[data-stat="team"] a[href]')
links_equipos = []
# Extraer los enlaces y agregarlos a la lista. Considera que debemos extraerlos como 'href', y no como texto.
for link in tbody_links:
    href = link['href']
    links_equipos.append(href)
```

```python
#A continuación procederemos a concatenar las url extraidas con la url inicial.
url_root = 'https://fbref.com/'
#Utilizamos urljoin, la cual es una función en Python que se utiliza para combinar dos partes de una URL
links_equipos = [urljoin(url_root,i) for i in links_equipos]
links_equipos
```

```python
#Vamos a utilizar la herramienta time para retrasar el tiempo entre solicitudes y asi evitar el error 429
import time
```

```python
#Ahora procederemos a extraer la data de cada uno de los links. 
#Data de fecha
date_matches = []
tiempo_de_pausa = 4 # Pausa de 4 segundos 

for link in links_equipos[0:96]:
# Realizar la solicitud GET a la página wed
    r_item=requests.get(link)
# Pausa entre solicitudes para evitar el error 429
    time.sleep(tiempo_de_pausa)  
# Parsear el contenido HTML con BeautifulSoup
    s_item=BeautifulSoup(r_item.text, 'html.parser')
# Encontrar el div con el id 'all_matchlogs'
    table_div = s_item.find('div', id='all_matchlogs')
# Encontrar el tbody dentro del div
    tbody = table_div.find('tbody')

# Encontrar los elementos th dentro del tbody y almacenarlos en la lista date_matches
    for th in tbody.select('th[data-stat="date"]'):
        date_match = th.get_text(strip=True)
        # Agregar el enlace al texto de la fecha
        date_match_with_link = f"{link}: {date_match}"
        date_matches.append(date_match_with_link)
# date_matches ahora contendrá las fechas junto con los enlaces
for item in date_matches:
    print(item)
```

```python
#Procedemos a separar la lista de elementos en 2 listas para posteriormente reeplazar los links por los nombres de los equipos.
enlaces = []
fechas = []

for links in date_matches:
    partes = links.split(": ")
    if len(partes) == 2:
        enlace, fecha = partes
        enlaces.append(enlace)
        fechas.append(fecha)
```
